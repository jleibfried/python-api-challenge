### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
cities_df = vacation_df[["Latitude", "Longitude"]]
humidity_df = vacation_df["Humidity"]
fig = gmaps.figure()
```

```python
heat_layer = gmaps.heatmap_layer(cities_df, weights=humidity_df, 
                                 dissipating=True, max_intensity=100,
                                 point_radius=1)

fig.add_layer(heat_layer)
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
# I am looking for a vacation to a place that is clear and cool for hiking
# Cloudiness < 10%
# Temperature about 10C 
# Wind is good but not too strong. 5-20 MPH seems about right
# Humidity in a cold place isn't that big of a deal so < 90

# trim for clouds
trimClouds_df = vacation_df[vacation_df.Cloudiness< 10]

# trim for temps
trimTemps_df = trimClouds_df[trimClouds_df.Temperature< 10]

# trim for wind
trimWindLow_df = trimTemps_df[trimTemps_df['Wind Speed']> 5]
trimWindHigh_df = trimWindLow_df[trimWindLow_df['Wind Speed']< 20]

# trim for humidity
trimTemps_df = trimWindHigh_df[trimWindHigh_df.Humidity< 90]

trimTemps_df.dropna()
trimTemps_df.head(20)
```
